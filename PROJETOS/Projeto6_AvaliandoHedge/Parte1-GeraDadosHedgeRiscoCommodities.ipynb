### <font color = darkred> Parte 1 - Gerando Dados Ficticios </font>

Com objetivo de entender o processo e depois aplicar em dados reais

```python
# Geração dos dados
# o seed permite que inicie o processo randomico, sempre com o mesmo padrao, mesmo que os valores mudem.
np.random.seed(0)
```

```python
# Número de dias
n = 365 
```

```python
# Preço inicial por barril de petróleo em dólares
preco_inicial = 50 
```

```python
# Volatilidade diária
volatilidade = 0.1  
```

```python
# drift_dsa diário
drift = 0.0002
```

```python
# Mudança percentual nos dados
mudanca_preco = np.random.normal(drift, volatilidade, n)
```

```python
# Gera os preços
precos = preco_inicial * np.exp(np.cumsum(mudanca_preco))
```

```python
# Cria o dataframe
df_preco = pd.DataFrame({'PrecoPetroleo': precos})
```

```python
# Salva em disco
df_preco.to_csv('preco_petroleo.csv', index = False)
```

```python
# Plot
plt.plot(df_preco['PrecoPetroleo'])
plt.title('Preço do Peróleo Durante 365 Dias')
plt.xlabel('Dias')
plt.ylabel('Preço em USD$')
plt.show()
```

**Volatilidade**  é  uma  medida  estatística  que  representa  a  dispersão  dos  retornos  de  um dado ativo ou mercado durante um período específico, indica o grau de variação de um ativo financeiro em um período de tempo. 

A volatilidade é frequentemente expressa  como  um  desvio  padrão  anualizado  dos  retornos  e  é  usada  tanto  em  modelos  de precificação de ativos como na avaliação do risco de mercado.

- **Volatilidade  Histórica:**  Baseia-se  em  dados  passados  e  calcula  o  desvio  padrão  dos retornos anteriores.

- **Volatilidade  Implícita:**  É  derivada  dos  preços  de  mercado  de  instrumentos  financeiros, como  opções,  e  reflete  as  expectativas  do  mercado  sobre  a  volatilidade  futura  do  ativo subjacente.

A volatilidade é usada na precificação de opções, na gestão de riscos e na construção de estratégias  de  trading.  Instrumentos  como  as  opções  de  compra  (call)  e  venda  (put)  são frequentemente usados para hedge em mercados voláteis.

O termo **"Drift"** refere-se à tendência direcional esperada do preço de um ativo ao longo do tempo, independentemente da sua volatilidade. 

Em modelos estocásticos de preço de ativos, como  o  modelo  de  Black-Scholes  para  a  precificação  de  opções,  o  drift  é  frequentemente representado como uma taxa de retorno constante que o investimento espera ganhar ao longo do tempo, ajustada ao risco.

- **Drift  Determinístico:**  É  uma  taxa  de  crescimento  constante  que  um  ativo  deve  seguir, geralmente expressa em termos de uma porcentagem anual.

- **Drift Estocástico:** Em modelos mais complexos, o drift pode ser uma função de diversas variáveis e pode mudar ao longo do tempo, tornando-se estocástico em vez de constante.

O conceito de drift é fundamental em modelos de precificação de ativos e simulações de Monte Carlo, onde é usado para modelar a trajetória esperada de preços de ativos.

Relação entre Volatilidade e Drift:Enquanto a volatilidade captura as flutuações de curto prazo no preço de um ativo, o drift reflete  a  expectativa  de  longo  prazo.  Ambos  os  conceitos  são  componentes-chave em modelos de precificação de ativos e são cruciais para a gestão de riscos e a construção de estratégias de investimento.
